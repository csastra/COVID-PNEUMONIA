```python
import glob
import os

# https://thispointer.com/python-get-list-of-files-in-directory-with-size/
# https://thispointer.com/python-get-file-size-in-kb-mb-or-gb-human-readable-format/

dir_name = 'D:/University/Y4S1/QBUS4934/data/extracted data/'

# Get a list of files (file paths) in the given directory 
list_of_files = filter( os.path.isfile,
                        glob.glob(dir_name + '*') )

# get list of ffiles with size
files_with_size = [ (file_path, os.stat(file_path).st_size) 
                    for file_path in list_of_files ]

# Iterate over list of tuples i.e. file_paths with size
# and print them one by one
for file_path, file_size in files_with_size:
    
    #convert file size to human-readable form (bytes to mb)
    size_kb = file_size/1024 # Bytes to KB
    size_mb = size_kb/1024 # KB to MB
    
    print(round(size_mb, 4), ' -->', os.path.basename(file_path)) 
```

```python
file_name_1 = []
old_size = []

for file_path, file_size in files_with_size:
    
    #get file name
    file_name_wtih_ext = os.path.basename(file_path)
    filename_without_ext = os.path.splitext(file_name_wtih_ext)[0]
    file_name_1.append(filename_without_ext)
    
    #get file size in MB
    size_mb = file_size/(1024*1024)
    old_size.append(size_mb)
```

```python
dir_name = 'D:/University/Y4S1/QBUS4934/data/new data/'

# Get a list of files (file paths) in the given directory 
list_of_files = filter( os.path.isfile,
                        glob.glob(dir_name + '*') )

# get list of ffiles with size
files_with_size = [ (file_path, os.stat(file_path).st_size) 
                    for file_path in list_of_files ]

# Iterate over list of tuples i.e. file_paths with size
# and print them one by one
for file_path, file_size in files_with_size:
    
    #convert file size to human-readable form (bytes to mb)
    size_kb = file_size/1024 # Bytes to KB
    size_mb = size_kb/1024 # KB to MB
    
    print(round(size_mb, 4), ' -->', os.path.basename(file_path)) 
```

```python
file_name_2 = []
new_size = []

for file_path, file_size in files_with_size:
    
    #get file name
    file_name_wtih_ext = os.path.basename(file_path)
    filename_without_ext = os.path.splitext(file_name_wtih_ext)[0]
    filename_without_ext_without_new = filename_without_ext.rsplit('_', 1)[0]
    file_name_2.append(filename_without_ext_without_new)
    
    #get file size in MB
    size_mb = file_size/(1024*1024)
    new_size.append(size_mb)
```

```python
a = set(file_name_1)
b = set(file_name_2)

if a == b:
    print("Lists are equal")
else:
    print("Lists are not equal")
print()

print(len(file_name_1))
print(len(old_size))
print(len(file_name_2))
print(len(new_size))
```

```python
import pandas as pd

df_old = pd.DataFrame(data = old_size, index = file_name_1, columns=['Old Size'])
df_old.head()
```

```python
df_new = pd.DataFrame(data = new_size, index = file_name_2, columns=['New Size'])
df_new.head()
```

```python
# https://stackoverflow.com/questions/40468069/merge-two-dataframes-by-index

df_merged = pd.merge(df_old, df_new, left_index=True, right_index=True) # inner join
# df_merged = df_new.join(df_old) # left join

df_merged
```

```python
df_merged.isna().sum()
```

```python
df_merged['Size Difference'] = df_merged['Old Size'] - df_merged['New Size'] 
df_merged
```

```python
import seaborn as sns

sns.set_theme(style="ticks")
sns.histplot(data=df_merged, x='Size Difference', bins=30)
```
